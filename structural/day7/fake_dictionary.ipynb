```python
import pandas as pd
import numpy as np

# Creating a fake dataset with "Dirty" data (Missing values)
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "salary": [50000, 60000, np.nan, 45000, 70000],  # NaN = Missing (Not a Number)
    "department": ["HR", "IT", "IT", np.nan, "HR"],
    "age": [25, 30, 35, 22, np.nan]
}

df = pd.DataFrame(data)
df.to_csv("employees.csv", index=False)

print("File 'employees.csv' created!")
```
