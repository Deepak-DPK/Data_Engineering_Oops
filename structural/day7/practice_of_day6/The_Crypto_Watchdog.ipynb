```python
import pandas as pd
```

```python
crypto = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "price_open": [45000, 46000, 45500, 48000, 47000],
    "price_close": [46000, 45500, 48000, 47000, 49000]
}
```

```python
df_crypto =pd.DataFrame(crypto)
df_crypto
```

```python
df_crypto["daily_change"] = df_crypto["price_close"] - df_crypto["price_open"]
df_crypto
```

```python
df_crypto[df_crypto["daily_change"] < 0]
```

```python
df_crypto[df_crypto["daily_change"] < 0].to_csv("bad_market_days.csv", index=False)
```
