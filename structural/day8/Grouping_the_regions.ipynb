```python
import pandas as pd

data = {
    "Region": ["North", "North", "South", "South", "East", "East", "North"],
    "Product": ["Apple", "Banana", "Apple", "Banana", "Apple", "Banana", "Apple"],
    "Sales": [100, 150, 200, 50, 120, 90, 80],
    "Profit": [30, 40, 60, 10, 40, 25, 20]
}

df = pd.DataFrame(data)
df
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
region_sales
```

```python
average_region_sales = df.groupby("Region")["rofit"].mean()
average_region_sales
```

```python
breakdown = df.groupby(["Region", "Product"])["Sales"].sum()
breakdown
```
