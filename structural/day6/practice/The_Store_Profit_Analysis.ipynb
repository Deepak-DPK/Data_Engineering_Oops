```python
import pandas as pd
```

```python
inventory = {
    "item": ["Drill", "Hammer", "Nails", "Saw", "Wrench"],
    "cost_price": [40, 5, 2, 15, 6],
    "selling_price": [60, 10, 5, 25, 8],
    "quantity_sold": [10, 50, 200, 15, 30]
}
```

```python
df_store =pd.DataFrame(inventory)
df_store
```

```python
df_store["profit_per_item"]=(df_store["selling_price"]- df_store["cost_price"])
```

```python
df_store
```

```python
df_store["total_revenue"] = (df_store["profit_per_item"] * df_store["quantity_sold"])
df_store
```

```python
df_store.to_csv("The_Store_Profit_Analysis.csv",index=False)
```
