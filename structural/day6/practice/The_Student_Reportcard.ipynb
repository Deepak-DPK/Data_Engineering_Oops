```python
import pandas as pd
```

```python
grades = {
    "Student": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "math": [85, 62, 90, 45, 88],
    "science": [92, 70, 95, 55, 82],
    "english": [78, 85, 88, 60, 90]
}
```

```python
df_grades= pd.DataFrame(grades)
df_grades
```

```python
df_grades["average"] =(df_grades["math"]+df_grades["science"]+df_grades["english"])/3

df_grades
```

```python
failing_students =df_grades[df_grades["average"] < 70]
failing_students
```
